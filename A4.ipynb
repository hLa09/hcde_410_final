```python
# Loading Crime dataset
# imports all necessary to write, and create graphs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the data in form of csv
data = pd.read_csv("Crime Data.csv")
# Provide the first few line of the dataset
data.head()
```

#### Counting number of crime offenses in each Offense_Category

After loading the data and preview what are the header, I then will count the crime categroy (type) to determine which crime are most commonly occur. As we can see the table above the crime types are list under the "Offense_Category", to get a quick look at the counts from the "Offense_Category" I will use the pandas function .value_counts to help me count the number of each category there is.

```python
# The following function will count each category that was offense from the year 2021 to 2023.
category = data['Offense_Category'].value_counts()
# Print the counts
category
```

With the help of .value_counts() function in pandas has made it easiser for me to count values in the crime category. This save me from having to type out each of the uniques category there is for crime. Though if you can see there are a label of DNE type, which is the left unknown category - where there is no label for the type of crime was reported.

#### Graphing the data of Offense Type of crimes

I will use the matplotlib.pyplot library to support me in graphing the data. I will use the function of .plot() and graph the data into a bar graph so that all the category is listed. To make the graph look a nicer to look at I will annotate them so that they show the total counts on top of the bar, this will allow us to distinct the differences of count in crime, as if there are no total counts label the graph show the category that has number lower than 50 look like 0 counts.

```python
# Plot the counts of Offense_Category into bar graph.
ax = category.plot(kind='bar', figsize = (25, 10), color='orange', fontsize = 12)
plt.title("Fig. # 1:  Crime occur in the City of Tacoma 2021 - 2023", fontsize = 16)
# Provides label for x-axis, and y-axis
plt.xlabel("Crime Category", fontsize = 14)
plt.ylabel("Number of crimes", fontsize = 14)

# Add counts on top of bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate('{}'.format(height),
                    xy=(p.get_x() + p.get_width() / 2, height),
                    fontsize = 12,
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
plt.show()
```

From the look of the graph we can see that the most crime counts was the Larceny/Theft Offenses, second to that was the Destruction/Damage/Vandalism, but what more surprising to me was there were count for Gambling Offenses. Although this graph only allow me to answer one question from the research question, next I will parse through the data again and creating a separate visual for the most crime occur in which neighborhood.

#### Counting numbers of crimes occurr in Day of the week

After looking at the crimes types and neighborhood I want to explore what day of the week occurr the most crime activities. This will allow me to grasp what day of the week I should be aware of, even though crime occurring will be many reason and number of crimes in the day of the week could not determine whehter or not crimes will always be more in the specific day of the week.

The following code is to help me read the recorded dates information present in the data, and convert them to an easier format that allow me to find number of crimes occur in the days of the week. We will be using datetime functino to format, and DatetimeIndex to change it to the index form for easier access.

```python
# Convert dates to pandas datetime format
data.Date_Occurred = pd.to_datetime(data.Date_Occurred, format='%m/%d/%Y')

# Set as index
data.index = pd.DatetimeIndex(data.Date_Occurred)
```

#### Graphing the data of Crimes occurr in days of the week of year 2021-2023

After converting the format, I then will create a list to indicate the days of the week and then plot them in bar-graph. I will still use the same format of editing the bar-graph for easier to visualized as above, annotate them so that we can see that actual total numbers and visualize the differences from the height of the bar.

```python
# Setting a list for days of the week
days = ['Monday','Tuesday','Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
# Plot the data for crime occur in days of week from 2021 - 2023
day_of_week = data.groupby([data.index.dayofweek]).size().plot(kind='bar', color='violet', figsize=(25, 10), fontsize=12)
# y-axis and x-xais label
plt.xlabel('Day of Week', fontsize = 14)
plt.xticks(np.arange(7), days, rotation=45)
plt.ylabel('Number of crimes', fontsize = 14)
plt.title('Fig. # 3: Number of crimes by days of week', fontsize = 16)

# Add counts on top of bars
for p in day_of_week.patches:
    height = p.get_height()
    day_of_week.annotate('{}'.format(height),
                    xy=(p.get_x() + p.get_width() / 2, height),
                    fontsize = 12,
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
plt.show()
```

As you can see from the above figure the day of the week that seem to occur more crime in the year of 2021 to the year of 2023 is the Friday, with the total crimes of 14709 activity. Which is about 1000 crimes different from other days in the week.

#### Graphing the data of Crimes occurr in month of years 2021-2023

After graphing the visualization, I am curious about what month occurr more crime? Therefore, I create a visualization for the number of crimes occurr in each months from the year 2021-2023. Thouhg I won't be going deep about why the people decide to choose that month to have the amount of crime as this project is not about finding why, but this project is about exploring the most common crime occurr, and which / when it seem to have the highest crime activities.

The following code using the same method of coding to find out what month has the highest crime activities, but this information will be present in the form of horizontal instead of vertical style like previous graphs.

```python
# Group data into months
ax= data.groupby([data.index.month]).size().plot(kind='barh', color='lightblue',figsize = (25,10), fontsize=12)

# y-axis label
plt.ylabel('Month', fontsize = 14)
# x-axis label
plt.xticks(rotation=360)
plt.xlabel('Number of crimes',  fontsize = 14)
# title label
plt.title('Fig. # 4: Number of crimes by month of years 2021-2023', fontsize=16)

for p in ax.patches:
    width = p.get_width()
    plt.text(1.025*p.get_width(), p.get_y()+0.55*p.get_height(),
             '{:1.2f}'.format(width),
             ha='center', va='center')

#show
plt.show()
```

From the look of the graph the month that have the highest number of crimes occurring is October, label in # 10 on the y-axis, with the total number of crimes of 8,788 incident. Second to the first is the January month, label in number 1 in the y-axis, with the total number of 8,245 incidents. If you look the graph as whole there is no significant difference between all months, we can see that every months there always activities out there that we should be watch out for. Since we have discussed the highest and second to highest, the month that has the lowest is in April with the total number of 6,996 activities. Even though it the lowest out of all month, but there still crime activities we should be aware of.
